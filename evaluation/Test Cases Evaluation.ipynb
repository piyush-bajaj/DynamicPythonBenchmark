```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data from CSV file
data = pd.read_csv('test.csv')
sorted_data = data.sort_values(by="project")

print("new project count : ", sorted_data.shape[0])

total_tests = sorted_data['total'].sum()
total_passed = sorted_data['passed'].sum()
total_failed = sorted_data['failed'].sum()
total_time = sorted_data['time'].sum()

max_time = sorted_data['time'].max()
min_time = sorted_data['time'].min()
max_tests = sorted_data['total'].max()
min_passed = sorted_data['passed'].min()

count_failed_0 = len(sorted_data[sorted_data['failed'] == 0])
count_failed_10 = len(sorted_data[sorted_data['failed'] <= 10])

count_total_passed_90 = len(sorted_data[(sorted_data['passed'] / sorted_data['total']) > 0.9 ])
count_total_passed_80 = len(sorted_data[(sorted_data['passed'] / sorted_data['total']) > 0.8 ])

count_run_50 = len(sorted_data[sorted_data['time'] > 50 ])
count_run_100 = len(sorted_data[sorted_data['time'] > 100 ])
count_run_200 = len(sorted_data[sorted_data['time'] > 200 ])

print("total tests : ", total_tests)
print("total tests passed: ", total_passed)
print("total tests faile: ", total_failed)
print("total test time : ", total_time)
print("max time : ", max_time)
print("min time : ", min_time)
print("max tests : ", max_tests)
print("percentage passed : ", total_passed/total_tests * 100)
print("min passed : ", min_passed)
print("projects with 0 failed tests : ", count_failed_0)
print("projects with 10 failed tests or less: ", count_failed_10)
print("projects with 90% passed test cases : ", count_total_passed_90)
print("projects with 80% passed test cases : ", count_total_passed_80)
print("run time greater than 50 seconds : ", count_run_50)
print("run time greater than 100 seconds : ", count_run_100)
print("run time greater than 200 seconds : ", count_run_200)
```
