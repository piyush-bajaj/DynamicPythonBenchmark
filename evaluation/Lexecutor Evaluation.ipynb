Below are the results with the validation of 5% of data points from all projects. (Random split)

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data from CSV file
data = pd.read_csv('validation_fine.csv')

# Extract the column names and data
columns = data.columns[1:]
values = data.iloc[:, 1:]

# Create a new figure and axis object
fig, ax = plt.subplots()

# Plot each line using a for loop
for i in range(len(columns)):
    ax.plot(data['Epoch'], values.iloc[:, i], label=columns[i])

# Set the x and y axis labels
ax.set_xlabel('Epoch')
ax.set_ylabel('Accuracy')

# Set the graph title
ax.set_title('Accuracy vs. Epoch - Random Split (fine-grained)')

# Show the legend
ax.legend()

# Show the graph
plt.show()
```
